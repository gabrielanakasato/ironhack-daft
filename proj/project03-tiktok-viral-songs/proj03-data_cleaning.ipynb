# Songs

## Import the dataset

```python
# Import the dataset
df_songs = pd.read_csv('cmc_tracks_data.csv')

# Select only the songs that were found in Spotify
df_songs = df_songs[df_songs.cmc_id != 'not-found']

# Check the result
df_songs
```

The results from the Chartmetric API became strings after exporting the dataframe. So, we need to convert them back to dictionaries.

```python
# Convert the Chartmetric columns from strings to dictionaries
df_songs.cmc_song_seach = df_songs.cmc_song_seach.apply(literal_eval)
df_songs.cmc_track_metadata = df_songs.cmc_track_metadata.apply(literal_eval)
df_songs.cmc_track_stats_tiktok = df_songs.cmc_track_stats_tiktok.apply(literal_eval)
```
